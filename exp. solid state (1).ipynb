```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
R = 8.314  # J/mol·K
Theta_E = 215  # Einstein temperature in K

# Temperature range
T/Theta_E== np.linspace(1, 1000, 1000)  # From 1 K to 1000 K

# Einstein model equation for heat capacity
x =Theta_E/T
C_V = 3 * R * (x**2 * np.exp(x)) / (np.exp(x) - 1)**2

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(T/215, C_V, label='Einstein Heat Capacity $C_V$', color='green')
plt.title('Heat Capacity C_V of Silver (Einstein Model)')
plt.xlabel('T/Theta_E')
plt.ylabel('Heat Capacity $C_V$ (J/mol·K)')
plt.grid(True)
plt.legend
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#constants
R= 8.314 #Ideal gas constant
Theta_E1= 428 #Einstein temperature for Al
Theta_E2= 105 #Einstein temperatue for Pb

#Temperature range
T=np.linspace(0, 500, 500)

#Einstein equation of Heat capacity 
x1=Theta_E1/T
C_V1 = 3 * R * (x1**2 * np.exp(x1)) / (np.exp(x1) - 1)**2

x2=Theta_E2/T
C_V2 = 3 * R * (x2**2 * np.exp(x2)) / (np.exp(x2) - 1)**2


#plotting
plt.figure(figsize=(10,5))
plt.plot(T, C_V1, color='blue') #Al
plt.plot(T, C_V2, color='green') #Pb
plt.title('comparision of heat capacity of Al and Pb') 
plt.grid(True) 
plt.xlabel('Temperature')
plt.ylabel('$C_v$: specific heat capacities, blue=Al, green=Pb')
plt.show              
```
